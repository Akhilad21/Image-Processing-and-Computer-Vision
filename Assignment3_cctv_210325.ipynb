```python
import cv2
import numpy as np

# Read the image from the current directory
input_image_number = int(input("Enter the image number (1, 2, 3, or 4): "))
image_path = f'cctv{input_image_number}.JPG'  # Assuming the images are in the same directory as the code

# Set the value of "a" based on the input image number
if input_image_number == 1:
    a = 200
elif input_image_number == 2:
    a = 0
elif input_image_number == 3:
    a = -10
elif input_image_number == 4:
    a = 75
else:
    print("Invalid image number. Please enter a value between 1 and 4.")
    exit()

# Read the image
img = cv2.imread(image_path, 0)

# Check if the image was successfully loaded
if img is None:
    print(f"Failed to load the image: {image_path}")
    exit()

# Perform contrast stretching
min_intensity = np.min(img)
max_intensity = np.max(img)
stretched_img = (img - min_intensity) * (255.0 / (max_intensity - min_intensity - a))

# Convert the image to uint8 format
stretched_img = stretched_img.astype(np.uint8)

# Save the enhanced image
enhanced_image_path = f'enhanced-cctv{input_image_number}.jpg'
cv2.imwrite(enhanced_image_path, stretched_img)

# Display the enhanced image
from PIL import Image
Image.open(enhanced_image_path)
```
